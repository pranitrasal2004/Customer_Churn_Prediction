```python
import pandas as pd
import numpy as np

np.random.seed(42)
n = 500
data = {
    "customer_id":range(1001,1001+n),
    "age":np.random.randint(18,65,n),
    "gender":np.random.choice(["Male","Female"],n),
    "tenure":np.random.randint(1,72,n),
    "monthly_charges":np.random.randint(300,2000,n),
    "total_charges":np.random.randint(1000,150000,n),
    "contract_type":np.random.choice(["Month-to-month","One year"],n),
    "internet_service":np.random.choice(["Yes", "No"], n),
    "payment_method": np.random.choice(["Credit Card", "Debit Card", "UPI"], n),
    "support_calls": np.random.randint(0, 10, n),
    "churn": np.random.choice(["Yes", "No"], n)
}
df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt
#age distribution
plt.hist(df['age'],bins=10)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
plt.hist(df['monthly_charges'],bins=10)
plt.xlabel('Monthly Charge')
plt.ylabel('Frequency')
plt.title('Monthly charge Distribution')
plt.show()
```

```python
#Tenure
plt.boxplot(df['tenure'])
plt.ylabel("Tenure(Months)")
plt.title("Tenure Boxplot")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
categorical_cols=['gender','contract_type','internet_service','churn','payment_method']

for col in categorical_cols:
    df[col]=le.fit_transform(df[col])
```
